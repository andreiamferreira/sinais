```python
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot') # aplica um estilo aos plots, que será aplicado a partir de agora automaticamente.
plt.rcParams['figure.figsize'] = [14, 5] #estabelece o tamanho da figura (FAÇA O AJUSTE DE ACORDO COM O TEu MONITOR)
```

# Construindo um sinal contínuo

```python
#Usando a função linspace:
# Linspace retorna números espaçados de modo uniforme em um intervalo. Dessa forma, dado um ponto inicial e de parada,
# assim como a quantidade de valores, linspace irá distribuí-los UNIFORMEMENTE para você em uma matriz NumPy.

t = np.linspace(0, 20, 10000)  # np.linspace(start, stop, num) -> valor inicial, final, quantidade ->  cria um array t com 10000 valores igualmente espaçados entre 0 e 20.
c_continuo = np.cos(t) * np.exp(-0.1 * t)

# np.cos(t) calcula o cosseno de cada valor no array t. Isso gera um array com os valores do cosseno para cada valor de t.
# np.exp(-0.1 * t) calcula a função exponencial 𝑒−0.1𝑡 para cada valor em t. Isso gera um array com valores que decaem exponencialmente com a mesma variável t.


# t e c: cada valor do cosseno é multiplicado pelo valor exponencial correspondente. Isso resulta em um array c_continuo que é uma função oscilatória (cosseno) com um fator de decaimento exponencial.


plt.plot(t, c_continuo)  # x, y
# gráfico da função c_continuo em função de t, mostrando como a função cosseno decaí em relação ao tempo.
```

# Construindo um sinal discreto

```python
n = np.linspace(0, 20, 21)  # [0, 1, 2, ..., 19, 20]

# calcula a função para os valores discretos em n (mesma formula do contínuo)
x_discreto = np.cos(n) * np.exp(-0.1*n)

plt.stem(n, x_discreto, use_line_collection=True)  #  cria um gráfico de stem/hastes -> n [e eixo x, x_discreto é y]

# plot de linhas/externo
plt.plot(t, c_continuo, '--')
```

# Exercício 1

Amostrando uma sinal com uma frequência de 2 Hz.

```python
n = np.linspace(0, 40, 41) # [0, 1, 2, ..., 39, 40] -> índices de amostragem
t = np.linspace(0, 20, 100)  # usado para calcular a função contínua no intervalo contínuo de tempo

# Calcula o sinal contínuo para cada valor em t
# função cosseno com decaimento exponencial
x_continuo = np.cos(t) * np.exp(-0.1*t)

fs = 2        #  frequência de amostragem = 2Hz
Ts = 1 / fs   # período de amostragem = .s

# Calcula o sinal amostrado para cada valor em n, onde n são os índices de amostragem.
sinal_discreto = np.cos(n * Ts) * np.exp(-0.1 * n * Ts)
# n * Ts dá os instantes de amostragem no tempo para os valores de n
# np.cos(n * Ts) calcula o cosseno nesses instantes
# np.exp(-0.1 * n * Ts) calcula o fator de decaimento exponencial correspondente

plt.stem(n/fs, sinal_discreto, use_line_collection=True)  # n / fs converte os índices de amostragem n para o tempo real, dado que a frequência de amostragem é 2 Hz. Portanto, n / fs vai de 0 a 20 segundos (com 41 amostras)
plt.plot(t, sinal_discreto, '--')
```

# Sinais Periódicos e Não-Periódicos
Deslocando um sinal periódico na horizontal (mexendo na FASE) por um número definido por  b .

```python
t = np.linspace(0, 5 * 2 * np.pi, 1000)

#b = 0.5* 2 * np.pi
b = 0.5* np.pi

x1 = np.sin(t)
x2 = np.sin(t + np.abs(b))
#x2=  np.cos(t)

plt.subplot(1,2,1)
plt.plot(t, x1)
plt.subplot(1,2,2)
plt.plot(t, x2)
```

# Deslocando um sinal não-periódico no tempo por um número de períodos  b:

```python
b = 5*np.pi  # Define o deslocamento de fase b como 0.5π (ou seja, 90 graus).

t = np.linspace(0, 5 * 2*np.pi, 1000)  # stop é 5×2𝜋  Isso significa que t vai de 0 até 10π (5 períodos completos de uma onda de seno)
# num é 1000, então o array t terá 1000 valores igualmente espaçados entre 0 e 10π. Isso é suficiente para mostrar vários períodos da função seno suavemente

# np.sin(t) -> Calcula o valor da função seno para cada valor em t. Esse é o sinal original sem deslocamento de fase
y_original =  np.sin(t) * np.exp(-0.1*t)

# Calcula o valor da função seno para cada valor em t, mas com um deslocamento de fase adicionado.
# np.abs(b) -> garante que b seja sempre positivo
# t + np.abs(b) -> desloca o sinal para a direita no eixo x. No caso do seno, isso é equivalente a uma mudança na fase da onda
y_deslocado =  np.sin(t + np.abs(b)) * np.exp(-0.1*(t + np.abs(b)))

plt.subplot(1,2,1)
plt.plot(t, y_original)
plt.ylim(-1.5, 1.5)
plt.subplot(1,2,2)
plt.plot(t, y_deslocado)
plt.ylim(-1.5, 1.5)
```
